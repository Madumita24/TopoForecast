```python
#3b - Models


import torch.nn as nn

class BiLSTMBase(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=128, num_layers=1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                            batch_first=True, bidirectional=True)
        self.head = nn.Linear(hidden_dim * 2, 1)

    def forward(self, x):
        out, _ = self.lstm(x)          # (B, T, 2H)
        h_last = out[:, -1, :]         # (B, 2H)
        return self.head(h_last)       # (B, 1)

class BiLSTMTopoAug(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=128, num_layers=1, topo_dim=3):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                            batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.Linear(hidden_dim * 2 + topo_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 1)
        )

    def forward(self, x, phi):
        out, _ = self.lstm(x)
        h_last = out[:, -1, :]
        z = torch.cat([h_last, phi], dim=1)
        return self.head(z)
```

\paragraph{Relative Weight Magnitude Interpretation.}

Let the first linear layer of the topology-augmented model be written as:
\[
y = W \cdot x,
\]
where the input vector is a concatenation of latent and topological features:
\[
x = [h_{\text{LSTM}}, \ \phi_{\text{topo}}].
\]

We partition the weight matrix accordingly:
\[
W = [W_{\text{LSTM}}, \ W_{\text{topo}}].
\]

To assess whether the model utilizes topological inputs, we compute the mean absolute weight magnitude:
\[
\bar{w}_{\text{LSTM}} = \mathbb{E}[|W_{\text{LSTM}}|], \quad
\bar{w}_{\text{topo}} = \mathbb{E}[|W_{\text{topo}}|].
\]

In our experiment:
\[
\bar{w}_{\text{LSTM}} \approx 0.058, \quad
\bar{w}_{\text{topo}} \approx 0.034,
\]
yielding the ratio:
\[
\frac{\bar{w}_{\text{topo}}}{\bar{w}_{\text{LSTM}}} \approx 0.58.
\]

This indicates that the topological features have weight magnitudes of the same order as the latent features, and are therefore not negligible. In other words, the model assigns meaningful weight to topological inputs rather than ignoring them.

\paragraph{Limitations.}

However, weight magnitude alone does not imply causal or conditional importance. The effective contribution of a feature depends on both its weight and its distribution:
\[
\text{influence} \sim W \cdot x.
\]
Thus, even non-zero weights may correspond to features used only as global biases or weak priors.

Therefore, weight-based analysis must be complemented with permutation tests to determine whether topological features are used as sample-specific conditioning signals.

## Reload dataset

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
CSV_PATH = r"../MSEA-Res_database_v2/Dynamic_components/Level2/Level2_5140_1974_Sirikit.csv"
#MSEA-Res_database_v2/Dynamic_components/Level2/Level2_5140_1974_Sirikit.csv
df = pd.read_csv(CSV_PATH)
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").dropna(subset=["Storage_km3"]).reset_index(drop=True)

print("Data shape:", df.shape)
print(df.head())
print(df.tail())
```

```python

# ----------------------------
# 2) Temporal split (train past, test future)
# ----------------------------
cutoff = pd.Timestamp("2016-01-01")  # same idea as before; change if you want
train_df = df[df["Date"] < cutoff].copy()
test_df  = df[df["Date"] >= cutoff].copy()

print("\nTrain:", train_df["Date"].min(), "→", train_df["Date"].max(), "n =", len(train_df))
print("Test :", test_df["Date"].min(),  "→", test_df["Date"].max(),  "n =", len(test_df))
```

```python

# ----------------------------
# 3) Scale using TRAIN only
# ----------------------------
scaler = StandardScaler().fit(train_df[["Storage_km3"]])
train_s = scaler.transform(train_df[["Storage_km3"]]).flatten()
test_s  = scaler.transform(test_df[["Storage_km3"]]).flatten()

# ----------------------------
# 4) Make supervised windows
# ----------------------------
def make_windows(series, lookback=180, horizon=1, stride=1):
    """
    series: 1D array
    returns:
      X: [N, lookback, 1]
      y: [N] if horizon=1 else [N, horizon]
    """
    X, y = [], []
    for i in range(0, len(series) - lookback - horizon + 1, stride):
        X.append(series[i:i+lookback])
        y.append(series[i+lookback:i+lookback+horizon])
    X = np.asarray(X)[:, :, None]
    y = np.asarray(y).squeeze()
    return X, y

LOOKBACK = 180
HORIZON  = 1

X_train, y_train = make_windows(train_s, lookback=LOOKBACK, horizon=HORIZON)
X_test,  y_test  = make_windows(test_s,  lookback=LOOKBACK, horizon=HORIZON)

print("\nWindowed shapes:")
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_test :", X_test.shape,  "y_test :", y_test.shape)
```

## Implementation of RMSE Evaluation

```python
import numpy as np
import torch

@torch.no_grad()
def rmse_aug_1day(model, X_np, phi_np, y_np, device, batch_size=256):
    model.eval()
    X = torch.tensor(X_np, dtype=torch.float32)
    P = torch.tensor(phi_np, dtype=torch.float32)

    preds = []
    for i in range(0, len(X), batch_size):
        xb = X[i:i+batch_size].to(device)
        pb = P[i:i+batch_size].to(device)
        yhat = model(xb, pb).detach().cpu().numpy()   # [B, 1] or [B]
        preds.append(yhat)

    pred = np.vstack(preds)

    # flatten safely (important for 1-day)
    pred = pred.reshape(-1)
    y_np = y_np.reshape(-1)

    assert pred.shape == y_np.shape, f"{pred.shape} vs {y_np.shape}"

    return float(np.sqrt(np.mean((pred - y_np) ** 2)))
```

```python
# 1) true RMSE
rmse_true = rmse_aug_1day(aug_best, X_test, phi_test_n, y_test, device)
print("RMSE (true phi):", rmse_true)

# 2) shuffled RMSEs
rng = np.random.default_rng(0)
n_perm = 20
rmse_perm = []

for k in range(n_perm):
    perm = rng.permutation(len(phi_test_n))
    phi_shuf = phi_test_n[perm]
    rm = rmse_aug_1day(aug_best, X_test, phi_shuf, y_test, device)
    rmse_perm.append(rm)

rmse_perm = np.array(rmse_perm)

print("RMSE (shuffled phi) mean±std:", rmse_perm.mean(), rmse_perm.std())
print("RMSE increase (mean):", rmse_perm.mean() - rmse_true)
print("All shuffled RMSEs:", rmse_perm)
```

```python
rmse_drop = []

for j in range(phi_test_n.shape[1]):
    phi_one = phi_test_n.copy()
    rng = np.random.default_rng(123 + j)
    phi_one[:, j] = phi_one[rng.permutation(len(phi_one)), j]
    rm = rmse_aug_1day(aug_best, X_test, phi_one, y_test, device)
    rmse_drop.append(rm)

rmse_drop = np.array(rmse_drop)

print("RMSE true:", rmse_true)
print("RMSE when shuffling each topo feature:", rmse_drop)
print("Increase per topo feature:", rmse_drop - rmse_true)
```

```python
# shuffle features 0 and 2 together
phi_both = phi_test_n.copy()
rng = np.random.default_rng(999)
perm = rng.permutation(len(phi_both))
phi_both[:, [0, 2]] = phi_both[perm][:, [0, 2]]

rmse_both = rmse_aug_1day(aug_best, X_test, phi_both, y_test, device)
print("RMSE shuffling features 0 and 2 together:", rmse_both)
print("Increase:", rmse_both - rmse_true)
```

```python
print("Relative increase (%):", (rmse_drop - rmse_true) / rmse_true * 100)
```

\paragraph{Feature-wise permutation analysis.}

While full permutation of the topological feature vector produces negligible change in RMSE, feature-wise perturbation reveals a more nuanced structure. Specifically, shuffling individual features results in:

\[
\Delta \mathrm{RMSE} \approx [0.0305,\,-0.0001,\,+0.0307],
\]

indicating that two of the three retained zigzag descriptors contribute substantially to predictive performance, while the third is largely redundant.

Furthermore, jointly shuffling the two dominant features yields a smaller but still positive degradation:

\[
\Delta \mathrm{RMSE} \approx 0.00138,
\]

suggesting that the model relies on their combined structure.

\paragraph{Interpretation.}

These results indicate that topological features are neither ignored nor used purely as independent conditional signals. Instead, the model appears to exploit the internal structure and joint distribution of the topological feature vector. This explains why full permutation across samples does not strongly degrade performance, while selective perturbation of individual features reveals significant dependence.

Overall, the findings suggest that topology acts as a structured low-dimensional signal, contributing to prediction through feature interactions rather than strict sample-wise alignment.
